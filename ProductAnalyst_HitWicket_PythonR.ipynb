```
#Q1

import pandas as pd
data = {
    'Class': [4, 4, 5, 1, 1, 2, 5, 2],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male'],
    'Values': [10, 3, 1, 5, 7, 2, 5, 10]
}

df = pd.DataFrame(data)
print(df)
```

```
#Q2
df_sorted = df.sort_values(by=['Class', 'Values'], ascending=[True, False])
print(df_sorted)
```

```
#Q3
df_grouped = df.groupby(['Class', 'Gender'])['Values'].nunique().reset_index(name='Unique_Values_Count')
print(df_grouped)
```

```
#Q4
import matplotlib.pyplot as plt
plt.scatter(df['Class'], df['Values'], c=df['Gender'].map({'Male': 'blue', 'Female': 'red'}))
plt.xlabel('Class')
plt.ylabel('Values')
plt.title('Class vs Values Scatter Plot')
plt.show()
```

```
#Q5
df_pivot = df.pivot_table(index='Class', columns='Gender', values='Values', aggfunc='sum').reset_index()
df_pivot = df_pivot.fillna(0)
df_pivot.columns.name = None
df_pivot.columns = ['Class', 'Female', 'Male']
print(df_pivot)
```

```
#Q6
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
table_to_plot = pd.DataFrame({
    'Category': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'x_variable': [1750, 950, 850, 1160, 350, 1450, 1550, 800, 850],
    'y_variable': [1850, 1150, 1000, 1240, 450, 1550, 1600, 900, 850]
})
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
for i, category in enumerate(sorted(table_to_plot['Category'].unique())):
    data = table_to_plot[table_to_plot['Category'] == category]
    ax = axes[i // 3, i % 3]
    sns.scatterplot(data=data, x='x_variable', y='y_variable', ax=ax)
    ax.set_title(f'Category {category}')
    ax.set_xlim(800, 2000)
    ax.set_ylim(800, 2000)
plt.tight_layout()
plt.show()
```

```
#Q7
import pandas as pd
table_to_plot = pd.DataFrame({
    'Category': [1, 1, 2],
    'x_variable': [1900, 1590, 995],
    'y_variable': [1875, 1770, 1205]
})
x_bins = pd.cut(table_to_plot['x_variable'], bins=[800, 1200, 1600, 2000], labels=['(800,1200)', '(1200,1600)', '(1600,2000)'])
y_bins = pd.cut(table_to_plot['y_variable'], bins=[800, 1200, 1600, 2000], labels=['(800,1200)', '(1200,1600)', '(1600,2000)'])
table_to_plot['x_range'] = x_bins
table_to_plot['y_range'] = y_bins
print(table_to_plot)
```
